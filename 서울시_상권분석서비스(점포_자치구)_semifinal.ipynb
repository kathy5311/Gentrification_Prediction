```python
import pandas as pd
```

[서울시 상권분석서비스(점포-자치구)] 데이터

```python
df=pd.read_csv("서울시 상권분석서비스(점포-자치구).csv", encoding='cp949')
df
```

```python
df.head(10)
```

```python
df1=df.loc[:, ['기준_년분기_코드','자치구_코드_명','서비스_업종_코드_명','점포_수','개업_점포_수','폐업_점포_수','프랜차이즈_점포_수']]
df1
```

```python
df1['서비스_업종_코드_명'].unique()
```

```python
# 서비스업종코드명에서 음식료품 관련한 것만 저장
df1 = df1[df1['서비스_업종_코드_명'].isin(['양식음식점', '커피-음료', '한식음식점', '패스트푸드점', '제과점', '분식전문점', '일식음식점', '중식음식점', '호프-간이주점', '치킨전문점'])]

#기준_년분기_코드를 년도로만 변경
df1['기준_년분기_코드']=df1['기준_년분기_코드'].astype(str).str[:4]
df1['기준_년분기_코드']=df1['기준_년분기_코드'].astype(int)
```

```python
df1.info()
```

```python
df1
```

```python
df1 = df1[df1['기준_년분기_코드'].isin([2019,2023])]
df1
```

칼럼 정리 (점포수->개인점포수, 각 칼럼의 '_'빼기 & 전제 점포, 프랜차이즈비율 열 추가)

```python
df1.rename(columns = {'점포_수':'개인점포수', '기준_년분기_코드':'기준년분기코드', '자치구_코드_명':'자치구코드명', '서비스_업종_코드_명':'서비스업종코드명', '개업_점포_수':'개업점포수', '폐업_점포_수':'폐업점포수', '프랜차이즈_점포_수':'프랜차이즈점포수'}, inplace=True)
df1
```

```python
df1['전체점포수'] = df1['프랜차이즈점포수'] + df1['개인점포수']
df1['개인점포-프랜차이즈'] = df1['프랜차이즈점포수'] / df1['개인점포수']
df1
```

```python
f_df1 = df1[df1['개인점포-프랜차이즈']>=0.4]
#f_df1.head(10)
f_df1['서비스업종코드명'].unique()
#f_df1.to_excel('개인점포-프랜차이즈.xlsx', encoding='CP949')
```

```python
df1['전체-프랜차이즈'] = df1['프랜차이즈점포수'] / df1['전체점포수']
df1
```

```python
fran_df1 = df1[df1['전체-프랜차이즈']>=0.4]
fran_df1['서비스업종코드명'].unique()
#fran_df1.head(10)
#fran_df1.to_excel('전체-프랜차이즈.xlsx', encoding='CP949')
```

```python
df_19=fran_df1[fran_df1['기준년분기코드']==2019]
df_19
```

```python
df_19.groupby(['자치구코드명','서비스업종코드명']).agg(['sum','mean'])

```

#2023

```python
df_23=fran_df1[fran_df1['기준년분기코드']==2023]
df_23
```

```python
df_23.groupby(['자치구코드명','서비스업종코드명']).agg(['sum','mean'])
```

#Merge

```python
group_merge=pd.merge(df_19, df_23, on=['자치구코드명','서비스업종코드명'], how='inner')
group_merge
```

#개업율 구하기

```python
group_merge['개업율']=(group_merge['개업점포수_y']-group_merge['개업점포수_x'])/group_merge['개업점포수_x']
group_merge
```

#폐업율

```python
group_merge['폐업율']=(group_merge['폐업점포수_y']-group_merge['폐업점포수_x'])/group_merge['폐업점포수_x']
group_merge
```

#개인증감율

```python
group_merge['개인점포_증감율']=((group_merge['개인점포수_y']/group_merge['전체점포수_y'])-(group_merge['개인점포수_x']/group_merge['전체점포수_x']))/(group_merge['개인점포수_x']/group_merge['전체점포수_x'])
group_merge
```

#프렌차이즈

```python
group_merge['프랜차이즈_증감율']=(group_merge['전체-프랜차이즈_y']-group_merge['전체-프랜차이즈_x'])/group_merge['개인점포-프랜차이즈_x']
group_merge
```
